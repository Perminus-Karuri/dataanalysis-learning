**Grouping and aggregation using .groupby() method**

The process of grouping rows of data together and performing aggregation on numerical columns is called grouping and aggregation. 
**Aggregation** is the process of grouping multiple values together to form a single summary value.

```python
import pandas as pd


df = pd.read_excel('Data_dump_10_rows.xlsx', sheet_name = 'Sheet1')
df
```

```python
df1 = df[['Gender', 'Country', 'Age']]
df1
```

**grouping by single column**

```python
by_country = df1.groupby(['Country'])
```

```python
by_country.mean(numeric_only = True)
```

```python
by_country.max() # maximum age in each country
```

```python
by_country.min() # minimum age in each country
```

```python
by_country.count() # count of people in each country
```

```python
by_country.std(numeric_only = True) # standard deviation
```

```python
by_country.describe()
```

Grouping by multiple columns

```python
by_country = df1.groupby(['Country', 'Gender'])
```

```python
by_country.count() # count each gender in a country
```

```python
by_country.mean() # mean age of every gender in the country
```

```python
by_country.min()
```

```python
by_country.describe()
```

```python
by_country.std()
```

```python
by_country.describe().transpose()
```

```python
by_country.describe().transpose()['United States'] # aggregation for a single country
```

**PIVOT TABLES**

A pivot table is used to summarise, sort, reorganize, group, count, total or average data stored in a table.
It allows grouping by any field(column), and it performs statistical calculations on them.
It allows us to transform columns into rows and rows into columns

```python
try:
    df = pd.read_excel('Excel_Sample.xlsx') # reading from an excel file
except Exception as e:
    print(f"Error occured: {e}")
```

```python
df
```

```python
df.pivot_table(values = 'Age', index = ['Country'], columns = ['Gender']) # creating a pivot table
```

```python
import numpy as np

df.pivot_table(values = 'Age', index = ['Country'], columns = ['Gender'], aggfunc = {'Age':['min','max', 'mean', 'sum']}) # creating a pivot table
```
