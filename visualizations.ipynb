```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# randomly generated values

df = pd.read_csv('Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
# print(plt.style.available)
plt.style.use('classic')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Ice cream ratings', xlabel = 'Daily ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar')
```

```python
df['Overall Rating'].plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked =  True)
```

```python
df.plot.barh()
```

```python
df.plot.scatter(x = 'Overall Rating', y = 'Texture Rating', s = 300, c = 'blue')
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = 'Overall Rating', figsize = (8,8))
```
