**Filtering by columns**

```python
df[df['Rank'] <= 10] # filtering by rank less than or equal to 10
```

```python
# Using the .isin() function to filter given a list of options

specific_countries = ['Bangladesh', 'Brazil']

df[df['Country'].isin(specific_countries)]
```

```python
# using .contains() function - returns countries with similar word in this case united

df[df['Country'].str.contains('United')]
```

**Filtering by rows**

```python
# setting index to country

df2 = df.set_index('Country')
df2
```

```python
df2.filter(items = ['Continent', 'CCA3'])
```

```python
df2.filter(items = ['Zimbabwe', 'Uganda'], axis = 0)
```

```python
df2.filter(like = 'United', axis = 0)
```

```python
df2.loc['Andorra']
```

```python
df2.iloc[90]
```

**Ordering**

```python
df
```

```python
df[df['Rank'] < 10].sort_values(by = 'Rank', ascending = False)
```

```python
df[df['Rank'] < 10].sort_values(by = ['Continent', 'Country'], ascending = [False, True])
```
