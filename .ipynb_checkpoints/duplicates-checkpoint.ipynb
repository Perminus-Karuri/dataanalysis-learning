```python
import pandas as pd

data = pd.read_csv('duplicates.csv')
```

```python
data
```

```python
# dropping first column

data.drop(columns='Unnamed: 0', inplace = True)
```

```python
# sorting by first name and year

data.sort_values(['First Name', 'Year'], inplace = True, ignore_index = True)
```

```python
data
```

```python
# Count of unique values in a columns

data['First Name'].nunique()
```

**Determining duplicated rows in DataFrame using .duplicated() function**

```python
data.duplicated()
```

```python
# count of duplicated rows in a DataFrame

data.duplicated().sum()
```

```python
# count of duplicated values in a column

data['First Name'].duplicated().sum()
```

```python
# count of duplicates with reagrds to values in specified columns

data.duplicated(subset = ['First Name', 'NO']).sum()
```

```python
# extracting duplicate rows from a DataFrame

data.loc[data.duplicated(), :]
```

**Dropping duplicate rows in a DataFrame uisng .drop_duplicate() function**

```python
data.drop_duplicates(inplace = True)
```

```python
data
```

```python
# Selecting the first duplicate row

data.drop_duplicates(subset = 'First Name', keep = 'first')
```

```python
# Selecting the last duplicate row

data.drop_duplicates(subset = 'First Name', keep = 'last')
```

```python
# Dropping all duplicate rows of a DataFrame

data.drop_duplicates(subset = ['First Name'], keep = False)  # to get unique first name values
```
